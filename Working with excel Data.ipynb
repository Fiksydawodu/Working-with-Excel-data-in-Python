## An Important lesson on importing Data with python

```python
location_data = pd.read_csv('Lending-company.csv', usecols = ['StringID', 'Location', 'Region'], index_col = 'StringID')
location_data.head()
```

```python
my_csv_data = pd.read_csv('Lending-company-single-column-data-Copy1.csv')
my_csv_data
```

```python
my_csv_data = pd.read_csv('Lending-company-single-column-data-Copy1.csv', sep = '\,') #, engine = 'python')
# without the engine parameter, python will fall back to it's own engine
my_csv_data.head()
```

```python
my_csv_data = pd.read_csv('Lending-company-single-column-data-Copy1.csv', sep = '\,', engine = 'python')
my_csv_data.head()
# But every row is taken as a string
```

```python
my_csv_data = pd.read_csv('Lending-company-single-column-data-Copy1.csv', delimiter = '\,', engine = 'python')
my_csv_data.head()
# Python takes both 'sep' and 'delimiter' paramters identically
```
